```python
# Merge on TransactionID using a left join to ensure all transactions are retained
merged_data = pd.merge(data1, data2, on='TransactionID', how='left')
merged_data.columns.tolist()
```

```python
print(merged_data.shape)  # Check the dimensions
print(merged_data.columns)  # Review all columns
print(merged_data.isnull().sum().sort_values(ascending=False))  # Check for missing values
```

```python
categorical_features = merged_data.select_dtypes(include=['object']).columns
merged_data[categorical_features] = merged_data[categorical_features].fillna('Missing')
numerical_features = merged_data.select_dtypes(include=['int64', 'float64']).columns
merged_data[numerical_features] = merged_data[numerical_features].fillna(merged_data[numerical_features].median())
```

```python
for c1, c2 in merged_data.dtypes.reset_index().values:
    if c2=='O':
        print(merged_data[c1].value_counts())
        print(f'\n > **Total Null Values:** {merged_data[c1].isnull().sum()}')
        print(f'\n > **Total Unique Values:** {len(merged_data[c1].unique())}')
        print('-' * 20 + '\n')
```

```python
from category_encoders import TargetEncoder
from sklearn.preprocessing import LabelEncoder, StandardScaler

low_card_features = ['ProductCD', 'card4', 'card6', 'M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9', 'DeviceType']
merged_data = pd.get_dummies(merged_data, columns=low_card_features, drop_first=True)

# Initialize Target Encoder
target_encoder = TargetEncoder(cols=['P_emaildomain', 'R_emaildomain', 'id_12', 'id_15', 'id_16', 'id_23', 
                                     'id_27', 'id_28', 'id_29', 'id_30', 'id_31', 'id_33', 'id_34', 
                                     'id_35', 'id_36', 'id_37', 'id_38', 'DeviceInfo'])

# Fit and transform
merged_data = target_encoder.fit_transform(merged_data, merged_data['isFraud'])

# Feature engineering (example: extracting time features)
merged_data['TransactionHour'] = merged_data['TransactionDT'] % 86400 // 3600
merged_data['TransactionDay'] = merged_data['TransactionDT'] // 86400
merged_data.drop('TransactionDT', axis=1, inplace=True)

# Feature scaling
scaler = StandardScaler()
scaled_features = ['TransactionAmt', 'TransactionHour', 'TransactionDay']
merged_data[scaled_features] = scaler.fit_transform(merged_data[scaled_features])
```

```python
merged_data.head(10)
```

```python
X = merged_data.drop(['TransactionID', 'isFraud'], axis=1)
y = merged_data['isFraud']
```

```python
from imblearn.over_sampling import SMOTE

# Reduce the number of synthetic samples generated by setting the sampling strategy.
smote = SMOTE(sampling_strategy=0.5, random_state=42)  # Adjust 0.5 as needed
X_resampled, y_resampled = smote.fit_resample(X, y)
X_resampled
```

```python
from imblearn.under_sampling import RandomUnderSampler

# After SMOTE, perform random undersampling
rus = RandomUnderSampler(sampling_strategy=1.0, random_state=42)  # Set desired ratio here
X_resampled, y_resampled = rus.fit_resample(X_resampled, y_resampled)
X_resampled
```

```python
import pandas as pd

# Assuming X_resampled and y_resampled are the features and target arrays
data_resampled = pd.concat([X_resampled, y_resampled], axis=1)

# Downsample to 200,000 rows
data_resampled_downsampled = data_resampled.sample(n=200000, random_state=42)
X_final = data_resampled_downsampled.drop('isFraud', axis=1)  # Replace 'target_column' with actual target column name
y_final = data_resampled_downsampled['isFraud']
column_list = data_resampled_downsampled.columns.tolist()
print(column_list)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Fit Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_final, y_final)

# Get feature importances
feature_importances = pd.Series(rf_model.feature_importances_, index=X_final.columns)
feature_importances = feature_importances.sort_values(ascending=False)
print(feature_importances)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for the plot
sns.set(style='whitegrid')
# Create a bar plot for the target variable 'isFraud'
plt.figure(figsize=(6, 4))
sns.countplot(x=y_final, palette='viridis')

# Add title and labels
plt.title('Distribution of Fraud and Non-Fraud Transactions')
plt.xlabel('Transaction Type (0: Non-Fraud, 1: Fraud)')
plt.ylabel('Count')

# Show plot
plt.show()
X_final
```
